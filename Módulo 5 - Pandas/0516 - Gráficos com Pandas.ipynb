# Pandas b√°sico

```python
import pandas as pd

BASE = "dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)

df.head()
```

```python
df.info()
```

```python
df["mfr"].value_counts().plot(
    kind="pie",
    ylabel="Percentual",
    autopct="%.1f%%",
    title="Percentual por Fabricantes"
)
```

```python
df.groupby("mfr")["calories"].mean().plot(
    kind="bar",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="M√©dia de Calorias por Fabricante"
)
```

```python
df.groupby("mfr")["calories"].mean().plot(
    kind="line",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="M√©dia de Calorias por Fabricante"
)
```
